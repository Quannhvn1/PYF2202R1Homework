```python
import turtle
import math
# pen set up
pen = turtle.Turtle()
pen.hideturtle()
pen.pensize(5)
pen.speed(100)
# create basic shapes
    # rectangle/square
def draw_rectangle(width,height,color='black'):
    pen.fillcolor(color)
    pen.begin_fill()
    for i in range(2):
        pen.fd(width)
        pen.lt(90)
        pen.fd(height)
        pen.lt(90)
    pen.end_fill()
    # isosceles right triangle
def draw_isosceles_right_triangle(side,color='black'):
    pen.fillcolor(color)
    pen.begin_fill()
    pen.fd(side*(2**0.5))
    pen.lt(135)
    pen.fd(side)
    pen.left(90)
    pen.fd(side)
    pen.end_fill()
    pen.rt(225)
    # parallelogram with 2 sides and angle between
def draw_parallelogram(side1, side2, angle, color='black'):
    pen.fillcolor(color)
    pen.begin_fill()
    pen.fd(side1)
    pen.lt(angle)
    pen.fd(side2)
    pen.lt(180-angle)
    pen.fd(side1)
    pen.lt(angle)
    pen.fd(side2)
    pen.end_fill()
    # circle
def draw_circle(radius,color='black'):
    pen.fillcolor(color)
    pen.begin_fill()
    pen.circle(radius)
    pen.end_fill()
    # lines with coordinates
def draw_lines(lenght,color='black'):
    pen.fillcolor(color)
    pen.begin_fill()
    pen.fd(lenght)
    pen.end_fill()
    # move to new position
def move_to(h,v):
    pen.penup()
    pen.goto(h,v)
    pen.pendown()

def draw_isosceles_trapezoid(side1, side2, angle, color='black'):
    pen.fillcolor(color)
    pen.begin_fill()
    pen.fd(side1)
    pen.rt(angle)
    pen.fd((side2-side1)/2/math.cos(angle/180*math.pi))
    pen.rt(180-angle)
    pen.fd(side2)
    pen.rt(180-angle)
    pen.fd((side2-side1)/2/math.cos(angle/180*math.pi))
    pen.rt(angle)
    pen.end_fill()
def move_to(h,v):
    pen.penup()
    pen.goto(h,v)
    pen.pendown()
def draw_lines(lenght,color='black'):
    pen.fillcolor(color)
    pen.begin_fill()
    pen.fd(lenght)
    pen.end_fill()

pen.speed(200)
#Tree1:
move_to(-300,-150)
pen.pencolor('blue')
draw_isosceles_trapezoid(10, 60, 75, '#808080')
move_to(-350,-150)
pen.pencolor('#525f30')
draw_isosceles_right_triangle(80,'#525f30')
move_to(-335,-93.4314575)
draw_isosceles_right_triangle(60,'#7c9048')
move_to(-320,-51.0050506288)
draw_isosceles_right_triangle(40,'#c6d4a1')
#Tree2:
move_to(200, 40)
pen.pencolor('blue')
draw_isosceles_trapezoid(20, 120, 80, '#808080')
move_to(100,40)
pen.pencolor('#525f30')
draw_isosceles_right_triangle(160,'#525f30')
move_to(130,153.1370849898476)
draw_isosceles_right_triangle(120,'#7c9048')
move_to(160,237.9898987322333)
draw_isosceles_right_triangle(80,'#c6d4a1')
#Factory
pen.pencolor('#5b9bd5')
move_to(-200,-100)
draw_rectangle(300,120,'white')
move_to(-70,-100)
draw_rectangle(40,60,'white')
move_to(-50,80)
draw_rectangle(20,60,'white')
move_to(-200,20)
draw_isosceles_right_triangle(160,'white')
# Smoke
move_to(-30,160)
pen.left(90)
draw_lines(60)
move_to(-40,160)
draw_lines(40)

turtle.done()
```
