```python
import turtle
import math
# pen set up
pen = turtle.Turtle()
pen.hideturtle()
pen.pensize(5)
pen.speed(100)
# create basic shapes
    # rectangle/square
def draw_rectangle(width,height,color='black'):
    pen.fillcolor(color)
    pen.begin_fill()
    for i in range(2):
        pen.fd(width)
        pen.lt(90)
        pen.fd(height)
        pen.lt(90)
    pen.end_fill()
    # isosceles right triangle
def draw_isosceles_right_triangle(side,color='black'):
    pen.fillcolor(color)
    pen.begin_fill()
    pen.fd(side*(2**0.5))
    pen.lt(135)
    pen.fd(side)
    pen.left(90)
    pen.fd(side)
    pen.end_fill()
    # parallelogram with 2 sides and angle between
def draw_parallelogram(side1, side2, angle, color='black'):
    pen.fillcolor(color)
    pen.begin_fill()
    pen.fd(side1)
    pen.lt(angle)
    pen.fd(side2)
    pen.lt(180-angle)
    pen.fd(side1)
    pen.lt(angle)
    pen.fd(side2)
    pen.end_fill()
    # circle
def draw_circle(radius,color='black'):
    pen.fillcolor(color)
    pen.begin_fill()
    pen.circle(radius)
    pen.end_fill()
    # lines with coordinates
def draw_lines(lenght,color='black'):
    pen.fillcolor(color)
    pen.begin_fill()
    pen.fd(lenght)
    pen.end_fill()
    # move to new position
def move_to(h,v):
    pen.penup()
    pen.goto(h,v)
    pen.pendown()

def draw_isosceles_trapezoid(side1, side2, angle, color='black'):
    pen.fillcolor(color)
    pen.begin_fill()
    pen.fd(side1)
    pen.rt(angle)
    pen.fd((side2-side1)/2/math.cos(angle/180*math.pi))
    pen.rt(180-angle)
    pen.fd(side2)
    pen.rt(180-angle)
    pen.fd((side2-side1)/2/math.cos(angle/180*math.pi))
    pen.end_fill()
def move_to(h,v):
    pen.penup()
    pen.goto(h,v)
    pen.pendown()
def draw_lines(lenght,color='black'):
    pen.fillcolor(color)
    pen.begin_fill()
    pen.fd(lenght)
    pen.end_fill()

pen.speed(100)
#Tree1:
draw_isosceles_trapezoid(20, 80, 80, 'white')
draw_isosceles_right_triangle(160,color='black')
draw_isosceles_right_triangle(140,color='black')
draw_isosceles_right_triangle(120,color='black')


turtle.done()
```
