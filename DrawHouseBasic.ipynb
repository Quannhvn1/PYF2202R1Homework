```python
import turtle
# pen set up
pen = turtle.Turtle()
pen.hideturtle()
pen.pensize(5)
pen.speed(100)
# create basic shapes
    # rectangle/square
    # isosceles right triangle
    # parallelogram with 2 sides and angle between
    # circle
    # lines with coordinates
    # move to new position
    
#draw house
# move_to(-300,-200)
pen.penup()
pen.goto(-300,-200)
pen.pendown()
    #blue square
# draw_rectangle(90,90,'blue')
pen.fillcolor('blue')
pen.begin_fill()
for i in range(2):
    pen.fd(90)
    pen.lt(90)
    pen.fd(90)
    pen.lt(90)
pen.end_fill()
    #green rectangle
# move_to(-270,-200)
pen.penup()
pen.goto(-270,-200)
pen.pendown()
# draw_rectangle(30,45,'#90ef8f')
pen.fillcolor('#90ef8f')
pen.begin_fill()
for i in range(2):
    pen.fd(30)
    pen.lt(90)
    pen.fd(45)
    pen.lt(90)
pen.end_fill()
    #yellow parallelogram
# move_to(-158.0384758, -170)
pen.penup()
pen.goto(-158.0384758, -170)
pen.pendown()
pen.lt(90)
# draw_parallelogram(90, 60, 120,'yellow')
pen.fillcolor('yellow')
pen.begin_fill()
pen.fd(90)
pen.lt(120)
pen.fd(60)
pen.lt(60)
pen.fd(90)
pen.lt(120)
pen.fd(60)
pen.end_fill()
    # purpil triangle
# move_to(-300, -110)
pen.penup()
pen.goto(-300, -110)
pen.pendown()
pen.rt(30)
# draw_isosceles_right_triangle(63.63961031,'#c4189a')
pen.fillcolor('#c4189a')
pen.begin_fill()
pen.fd(63.63961031*(2**0.5))
pen.lt(135)
pen.fd(63.63961031)
pen.left(90)
pen.fd(63.63961031)
pen.end_fill()
    #orange parallelogram
# move_to(-210, -110)
pen.penup()
pen.goto(-210, -110)
pen.pendown()
pen.lt(165)
# draw_parallelogram(60, 63.63961031, 105,'orange')
pen.fillcolor('orange')
pen.begin_fill()
pen.fd(60)
pen.lt(105)
pen.fd(63.63961031)
pen.lt(75)
pen.fd(60)
pen.lt(105)
pen.fd(63.63961031)
pen.end_fill()
    #brown parallelogram
# move_to(-190, -155)
pen.penup()
pen.goto(-190, -155)
pen.pendown()
pen.lt(75)
# draw_parallelogram(20, 30, 60,'#a62a2a')
pen.fillcolor('#a62a2a')
pen.begin_fill()
pen.fd(20)
pen.lt(60)
pen.fd(30)
pen.lt(120)
pen.fd(20)
pen.lt(60)
pen.fd(30)
pen.end_fill()

#draw tree
    # brown rectangle
# move_to(200,-200)
pen.penup()
pen.goto(200,-200)
pen.pendown()
pen.lt(90)
# draw_rectangle(20,60,'#a62a2a')
pen.fillcolor('#a62a2a')
pen.begin_fill()
for i in range(2):
    pen.fd(20)
    pen.lt(90)
    pen.fd(60)
    pen.lt(90)
pen.end_fill()
    #green leaves
# move_to(160, -140)
pen.penup()
pen.goto(160, -140)
pen.pendown()
# draw_isosceles_right_triangle(70.71067812,'#90ef8f')
pen.fillcolor('#90ef8f')
pen.begin_fill()
pen.fd(70.71067812*(2**0.5))
pen.lt(135)
pen.fd(70.71067812)
pen.left(90)
pen.fd(70.71067812)
pen.end_fill()
# move_to(170, -90)
pen.penup()
pen.goto(170, -90)
pen.pendown()
pen.lt(135)
# draw_isosceles_right_triangle(56.56854249,'#90ef8f')
pen.fillcolor('#90ef8f')
pen.begin_fill()
pen.fd(56.56854249*(2**0.5))
pen.lt(135)
pen.fd(56.56854249)
pen.left(90)
pen.fd(56.56854249)
pen.end_fill()
# move_to(180, -50)
pen.penup()
pen.goto(180, -50)
pen.pendown()
pen.lt(135)
# draw_isosceles_right_triangle(42.42640687,'#90ef8f')
pen.fillcolor('#90ef8f')
pen.begin_fill()
pen.fd(42.42640687*(2**0.5))
pen.lt(135)
pen.fd(42.42640687)
pen.left(90)
pen.fd(42.42640687)
pen.end_fill()

#draw sun
    # black lines
# move_to(0,250)
pen.penup()
pen.goto(0,250)
pen.pendown()
pen.lt(135)
# draw_lines(100)
pen.fillcolor('black')
pen.begin_fill()
pen.fd(100)
pen.end_fill()
# move_to(50,300)
pen.penup()
pen.goto(50,300)
pen.pendown()
pen.rt(90)
# draw_lines(100)
pen.fillcolor('black')
pen.begin_fill()
pen.fd(100)
pen.end_fill()
# move_to(25,275)
pen.penup()
pen.goto(25,275)
pen.pendown()
pen.lt(45)
# draw_lines(70.71067812)
pen.fillcolor('black')
pen.begin_fill()
pen.fd(70.71067812)
pen.end_fill()
# move_to(25,225)
pen.penup()
pen.goto(25,225)
pen.pendown()
pen.lt(90)
# draw_lines(70.71067812)
pen.fillcolor('black')
pen.begin_fill()
pen.fd(70.71067812)
pen.end_fill()
    # yellow circle
pen.speed(1)
# move_to(67.67766953, 232.3223305)
pen.penup()
pen.goto(67.67766953, 232.3223305)
pen.pendown()
# draw_circle(25,'yellow')
pen.fillcolor('yellow')
pen.begin_fill()
pen.circle(25)
pen.end_fill()
# print(pen.circle.__doc__)
turtle.done()
```
