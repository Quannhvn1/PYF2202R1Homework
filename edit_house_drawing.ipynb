```python
import turtle
import math
# pen set up
pen = turtle.Turtle()
pen.hideturtle()
pen.pensize(5)
pen.speed(100)
# create basic shapes
    # rectangle/square
    # isosceles right triangle
    # parallelogram with 2 sides and angle between
    # circle
    # lines with coordinates
    # move to new position
    
#draw house
# move_to(-300,-200)
x1 = -300
y1 = -200
pen.penup()
pen.goto(x1,y1)
pen.pendown()
    #blue square
blue_side = 90
# draw_rectangle(90,90,'blue')
pen.fillcolor('blue')
pen.begin_fill()
for i in range(2):
    pen.fd(blue_side)
    pen.lt(90)
    pen.fd(blue_side)
    pen.lt(90)
pen.end_fill()
    #green rectangle
# move_to(-270,-200)
green_width = blue_side / 3
green_height = blue_side /2
x2 = x1 + green_width
y2 = y1
pen.penup()
pen.goto(x2 , y2)
pen.pendown()
# draw_rectangle(30,45,'#90ef8f')
pen.fillcolor('#90ef8f')
pen.begin_fill()
for i in range(2):
    pen.fd(green_width)
    pen.lt(90)
    pen.fd(green_height)
    pen.lt(90)
pen.end_fill()
    #yellow parallelogram
# move_to(-158.0384758, -170)
yellow_side1 = blue_side
yellow_side2 = 60
yellow_angle = 120
x3 = x1 + yellow_side1 + yellow_side2*(math.sin(math.pi/3))
y3 = y1 + yellow_side2*(math.cos(math.pi/3))
pen.penup()
pen.goto(x3, y3)
pen.pendown()
pen.lt(90)
# draw_parallelogram(90, 60, 120,'yellow')
pen.fillcolor('yellow')
pen.begin_fill()
pen.fd(yellow_side1)
pen.lt(yellow_angle)
pen.fd(yellow_side2)
pen.lt(180-yellow_angle)
pen.fd(yellow_side1)
pen.lt(yellow_angle)
pen.fd(yellow_side2)
pen.end_fill()
    # purpil triangle
# move_to(-300, -110)
x4 = x1 
y4 = y1 + blue_side
purpil_side = blue_side/(2**0.5)
purpil_hypotenuse = blue_side
pen.penup()
pen.goto(x4, y4)
pen.pendown()
pen.rt(30)
# draw_isosceles_right_triangle(63.63961031,'#c4189a')
pen.fillcolor('#c4189a')
pen.begin_fill()
pen.fd(purpil_hypotenuse)
pen.lt(135)
pen.fd(purpil_side)
pen.left(90)
pen.fd(purpil_side)
pen.end_fill()
    #orange parallelogram
# move_to(-210, -110)
x5 = x4 + blue_side
y5 = y4
orange_side1 = yellow_side2
orange_side2 = purpil_side
pen.penup()
pen.goto(x5 , y5)
pen.pendown()
pen.lt(165)
# draw_parallelogram(60, 63.63961031, 105,'orange')
pen.fillcolor('orange')
pen.begin_fill()
pen.fd(orange_side1)
pen.lt(105)
pen.fd(orange_side2)
pen.lt(75)
pen.fd(orange_side1)
pen.lt(105)
pen.fd(orange_side2)
pen.end_fill()
    #brown parallelogram
# move_to(-190, -155)
x6 = x5 + yellow_side2*(math.sin(math.pi/3))/3
y6 = y2 + green_height
brown_side1 = yellow_side2/3
brown_side2 = yellow_side1/3
pen.penup()
pen.goto(x6, y6)
pen.pendown()
pen.lt(75)
# draw_parallelogram(20, 30, 60,'#a62a2a')
pen.fillcolor('#a62a2a')
pen.begin_fill()
pen.fd(brown_side1)
pen.lt(60)
pen.fd(brown_side2)
pen.lt(120)
pen.fd(brown_side1)
pen.lt(60)
pen.fd(brown_side2)
pen.end_fill()

#draw tree
    # brown rectangle
# move_to(200,-200)
x7 = 200
y7 = y1
brown_width = 20
brown_height = 60
pen.penup()
pen.goto(x7,y7)
pen.pendown()
pen.lt(90)
# draw_rectangle(20,60,'#a62a2a')
pen.fillcolor('#a62a2a')
pen.begin_fill()
for i in range(2):
    pen.fd(brown_width)
    pen.lt(90)
    pen.fd(brown_height)
    pen.lt(90)
pen.end_fill()
    #green leaves
# move_to(160, -140)
leaf1 = 100
leaf1_side = 100/(2**0.5)
x8 = x7 - (leaf1 - brown_width)/2
y8 = y7 + brown_height
pen.penup()
pen.goto(x8 , y8)
pen.pendown()
# draw_isosceles_right_triangle(70.71067812,'#90ef8f')
pen.fillcolor('#90ef8f')
pen.begin_fill()
pen.fd(leaf1)
pen.lt(135)
pen.fd(leaf1_side)
pen.left(90)
pen.fd(leaf1_side)
pen.end_fill()
# move_to(170, -90)
leaf2 = 90
leaf2_side = 90/(2**0.5)
x9 = x7 - (leaf2 - brown_width)/2
y9 = y8 + leaf1/2
pen.penup()
pen.goto(x9, y9)
pen.pendown()
pen.lt(135)
# draw_isosceles_right_triangle(56.56854249,'#90ef8f')
pen.fillcolor('#90ef8f')
pen.begin_fill()
pen.fd(leaf2)
pen.lt(135)
pen.fd(leaf2_side)
pen.left(90)
pen.fd(leaf2_side)
pen.end_fill()
# move_to(180, -50)
leaf3 = 80
leaf3_side = 80/(2**0.5)
x9 = x7 - (leaf3 - brown_width)/2
y9 = y8 + leaf1 / 2 + leaf2 / 2
pen.penup()
pen.goto( x9, y9)
pen.pendown()
pen.lt(135)
# draw_isosceles_right_triangle(42.42640687,'#90ef8f')
pen.fillcolor('#90ef8f')
pen.begin_fill()
pen.fd(leaf3)
pen.lt(135)
pen.fd(leaf3_side)
pen.left(90)
pen.fd(leaf3_side)
pen.end_fill()

#draw sun
    # black lines
# move_to(0,250)
pen.penup()
pen.goto(0,250)
pen.pendown()
pen.lt(135)
# draw_lines(100)
pen.fillcolor('black')
pen.begin_fill()
pen.fd(100)
pen.end_fill()
# move_to(50,300)
pen.penup()
pen.goto(50,300)
pen.pendown()
pen.rt(90)
# draw_lines(100)
pen.fillcolor('black')
pen.begin_fill()
pen.fd(100)
pen.end_fill()
# move_to(25,275)
pen.penup()
pen.goto(25,275)
pen.pendown()
pen.lt(45)
# draw_lines(70.71067812)
pen.fillcolor('black')
pen.begin_fill()
pen.fd(70.71067812)
pen.end_fill()
# move_to(25,225)
pen.penup()
pen.goto(25,225)
pen.pendown()
pen.lt(90)
# draw_lines(70.71067812)
pen.fillcolor('black')
pen.begin_fill()
pen.fd(70.71067812)
pen.end_fill()
    # yellow circle
pen.speed(1)
# move_to(67.67766953, 232.3223305)
pen.penup()
pen.goto(67.67766953, 232.3223305)
pen.pendown()
# draw_circle(25,'yellow')
pen.fillcolor('yellow')
pen.begin_fill()
pen.circle(25)
pen.end_fill()
# print(pen.circle.__doc__)
turtle.done()
```
