```python
import turtle
# pen set up
pen = turtle.Turtle()
pen.hideturtle()
pen.pensize(5)
pen.speed(100)
# create basic shapes
    # rectangle/square
def draw_rectangle(width,height,color='black'):
    pen.fillcolor(color)
    pen.begin_fill()
    for i in range(2):
        pen.fd(width)
        pen.lt(90)
        pen.fd(height)
        pen.lt(90)
    pen.end_fill()
    # isosceles right triangle
def draw_isosceles_right_triangle(side,color='black'):
    pen.fillcolor(color)
    pen.begin_fill()
    pen.fd(side*(2**0.5))
    pen.lt(135)
    pen.fd(side)
    pen.left(90)
    pen.fd(side)
    pen.end_fill()
    # parallelogram with 2 sides and angle between
def draw_parallelogram(side1, side2, angle, color='black'):
    pen.fillcolor(color)
    pen.begin_fill()
    pen.fd(side1)
    pen.lt(angle)
    pen.fd(side2)
    pen.lt(180-angle)
    pen.fd(side1)
    pen.lt(angle)
    pen.fd(side2)
    pen.end_fill()
    # circle
def draw_circle(radius,color='black'):
    pen.fillcolor(color)
    pen.begin_fill()
    pen.circle(radius)
    pen.end_fill()
    # lines with coordinates
def draw_lines(lenght,color='black'):
    pen.fillcolor(color)
    pen.begin_fill()
    pen.fd(lenght)
    pen.end_fill()
    # move to new position
def move_to(h,v):
    pen.penup()
    pen.goto(h,v)
    pen.pendown()
    
#draw house
move_to(-300,-200)
    #blue square
# draw_rectangle(90,90,'blue')
pen.fillcolor(color)
    pen.begin_fill()
    for i in range(2):
        pen.fd(width)
        pen.lt(90)
        pen.fd(height)
        pen.lt(90)
    pen.end_fill()
    #green rectangle
move_to(-270,-200)
# draw_rectangle(30,45,'#90ef8f')
pen.fillcolor(color)
    pen.begin_fill()
    for i in range(2):
        pen.fd(width)
        pen.lt(90)
        pen.fd(height)
        pen.lt(90)
    pen.end_fill()
    #yellow parallelogram
move_to(-158.0384758, -170)
pen.lt(90)
draw_parallelogram(90, 60, 120,'yellow')
    # purpil triangle
move_to(-300, -110)
pen.rt(30)
draw_isosceles_right_triangle(63.63961031,'#c4189a')
    #orange parallelogram
move_to(-210, -110)
pen.lt(165)
draw_parallelogram(60, 63.63961031, 105,'orange')
    #brown parallelogram
move_to(-190, -155)
pen.lt(75)
draw_parallelogram(20, 30, 60,'#a62a2a')

#draw tree
    # brown rectangle
move_to(200,-200)
pen.lt(90)
# draw_rectangle(20,60,'#a62a2a')
pen.fillcolor(color)
    pen.begin_fill()
    for i in range(2):
        pen.fd(width)
        pen.lt(90)
        pen.fd(height)
        pen.lt(90)
    pen.end_fill()
    #green leaves
move_to(160, -140)
draw_isosceles_right_triangle(70.71067812,'#90ef8f')
move_to(170, -90)
pen.lt(135)
draw_isosceles_right_triangle(56.56854249,'#90ef8f')
move_to(180, -50)
pen.lt(135)
draw_isosceles_right_triangle(42.42640687,'#90ef8f')

#draw sun
    # black lines
move_to(0,250)
pen.lt(135)
draw_lines(100)
move_to(50,300)
pen.rt(90)
draw_lines(100)
move_to(25,275)
pen.lt(45)
draw_lines(70.71067812)
move_to(25,225)
pen.lt(90)
draw_lines(70.71067812)
    # yellow circle
pen.speed(1)
move_to(67.67766953, 232.3223305)
draw_circle(25,'yellow')
# print(pen.circle.__doc__)
turtle.done()
```
